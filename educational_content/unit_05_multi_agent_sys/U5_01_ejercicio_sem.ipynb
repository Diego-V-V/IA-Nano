## Ejercicio de Extensión U5_01: Clasificador SEM (Proyecto Final)

Aplicamos los conceptos de **agentes, tools y warm-up** al dominio del Proyecto Final:
clasificación de morfologías en imágenes SEM (Nanopartículas vs Nanohilos).

### Objetivos del Ejercicio
1. Definir 2 tools específicas del dominio SEM (`get_model_metrics`, `interpret_gradcam`)
2. Registrar el dominio `sem_classification` en `context_loader.py`
3. Comparar un agente **sin warm-up** vs **con warm-up** para la misma pregunta
4. Justificar los hiperparámetros óptimos para el dominio

````python
# ============================================================
# SETUP: Conexión LLM via OpenRouter
# ============================================================
import os, sys, time
from pathlib import Path
from dotenv import load_dotenv

load_dotenv(override=True)
os.environ["OPENROUTER_API_KEY"] = os.environ.get(
    "OPENROUTER_API_KEY",
    "TU_OPENROUTER_API_KEY"
)

from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_tool_calling_agent, AgentExecutor
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

OPENROUTER_API_KEY = os.environ["OPENROUTER_API_KEY"]
MODELS_FALLBACK = ["openai/gpt-4o-mini", "anthropic/claude-3-haiku", "google/gemini-2.5-flash"]

llm = None
OPENROUTER_MODEL = ""
for model_id in MODELS_FALLBACK:
    try:
        _c = ChatOpenAI(
            base_url="https://openrouter.ai/api/v1",
            api_key=OPENROUTER_API_KEY,
            model=model_id,
            temperature=0,
            default_headers={
                "HTTP-Referer": "https://github.com/antigravity-nano",
                "X-Title": "SEM Classifier Exercise",
            },
        )
        _c.invoke("OK")
        llm = _c
        OPENROUTER_MODEL = model_id
        print(f"LLM conectado: {model_id}")
        break
    except Exception as e:
        print(f"  [!] {model_id}: {e}")

if llm is None:
    raise RuntimeError("No se pudo conectar a ningun modelo.")
````

### Paso 1: Definición de Tools del Dominio SEM

Creamos 2 tools específicas para el clasificador SEM:
- **`get_model_metrics`**: Retorna las métricas reales del modelo ResNet-18
- **`interpret_gradcam`**: Interpreta los mapas de activación para cada morfología

````python
# ============================================================
# 1. TOOLS DEL DOMINIO SEM
# ============================================================

@tool
def get_model_metrics(model_id: str) -> str:
    """Obtiene las métricas de rendimiento de un modelo de clasificación SEM.
    Usa esta tool cuando te pregunten sobre Accuracy, F1-score o errores de un modelo."""
    if "resnet" in model_id.lower():
        return (
            "Métricas de ResNet-18 para SEM:\n"
            "- Accuracy: 99.07% (106/107 correctas)\n"
            "- F1-Score: 0.9903\n"
            "- Precision: 0.9907\n"
            "- Recall: 0.9907\n"
            "- AUC-ROC: 0.9943\n"
            "- Errores: 1 de 107 imágenes\n"
            "- Clases: Nanoparticles (0D), Nanowires (1D)\n"
            "- Dataset: NFFA-EUROPE SEM\n"
            "- Matriz de confusión: [[53, 0], [1, 53]]"
        )
    return f"Métricas para el modelo {model_id} no encontradas."

@tool
def interpret_gradcam(morfologia: str) -> str:
    """Interpreta el mapa de activación Grad-CAM para una morfología SEM específica.
    Usa esta tool para explicar qué partes de la imagen está mirando el modelo."""
    m = morfologia.lower()
    if "nanoparticula" in m or "0d" in m or "particle" in m:
        return (
            "Grad-CAM para Nanopartículas (0D):\n"
            "- Alta activación en los bordes esféricos y contornos redondeados\n"
            "- El modelo ignora el fondo liso del sustrato\n"
            "- Detecta la relación aspecto ~1:1 característica de partículas"
        )
    elif "nanohilo" in m or "nanowire" in m or "1d" in m:
        return (
            "Grad-CAM para Nanohilos (1D):\n"
            "- Fuerte activación a lo largo de las estructuras elongadas y lineales\n"
            "- Detecta patrones direccionales y alta relación aspecto (>10:1)\n"
            "- Se centra en los ejes longitudinales, no en los extremos"
        )
    return "Morfología no reconocida. Usa: nanopartícula, nanohilo, 0D, o 1D."

tools_sem = [get_model_metrics, interpret_gradcam]
print("Tools SEM definidas:", [t.name for t in tools_sem])
print()

# Probar tools directamente
print("=== Prueba directa de tools ===")
print(get_model_metrics.invoke({"model_id": "ResNet-18"}))
print()
print(interpret_gradcam.invoke({"morfologia": "nanopartículas 0D"}))
````

### Paso 3: Comparación Sin Warm-Up vs Con Warm-Up

Creamos dos agentes idénticos con los mismos tools, pero uno tiene un prompt genérico y el otro el prompt especializado de SEM. Ejecutamos la misma pregunta en ambos.

````python
# ============================================================
# 3. AGENTE SIN WARM-UP (Genérico)
# ============================================================
prompt_sin = ChatPromptTemplate.from_messages([
    ("system", "Eres un asistente de IA útil. Usa las tools disponibles si es necesario."),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])
agent_sin = create_tool_calling_agent(llm, tools_sem, prompt_sin)
executor_sin = AgentExecutor(
    agent=agent_sin, tools=tools_sem,
    verbose=True, max_iterations=3, handle_parsing_errors=True,
)

pregunta = (
    "¿Cómo interpreta el modelo ResNet-18 las nanopartículas frente a los nanohilos "
    "y cuáles son sus métricas generales?"
)

print("=" * 70)
print("AGENTE SIN WARM-UP (prompt genérico)")
print("=" * 70)
resp_sin = executor_sin.invoke({"input": pregunta})
print("\nRESPUESTA FINAL:")
print(resp_sin["output"])
````

````python
# ============================================================
# 4. AGENTE CON WARM-UP (Especialista SEM)
# ============================================================
prompt_con = ChatPromptTemplate.from_messages([
    ("system", ctx_sem["system_context"]),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])
agent_con = create_tool_calling_agent(llm, tools_sem, prompt_con)
executor_con = AgentExecutor(
    agent=agent_con, tools=tools_sem,
    verbose=True, max_iterations=3, handle_parsing_errors=True,
)

print("=" * 70)
print("AGENTE CON WARM-UP (dominio sem_classification)")
print("=" * 70)
resp_con = executor_con.invoke({"input": pregunta})
print("\nRESPUESTA FINAL:")
print(resp_con["output"])
````

### Paso 4: Hiperparámetros Óptimos para el Dominio SEM

| Hiperparámetro | Valor | Justificación |
|---|---|---|
| `temperature` | **0.1** | Métricas científicas exactas (99.07% accuracy) requieren respuestas deterministas. Una T alta genera alucinaciones en los números. |
| `top_p` | **0.9** | Permite fluidez en explicaciones físicas pero corta tokens irrelevantes. |
| `frequency_penalty` | **0.3** | Evita repetición literal al comparar dos morfologías distintas. |
| `max_tokens` | **500** | Suficiente para análisis completo de métricas + Grad-CAM sin redundancia. |
