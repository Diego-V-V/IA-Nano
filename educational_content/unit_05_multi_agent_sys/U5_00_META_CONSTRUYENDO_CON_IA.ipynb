## 2. Antigravity como Co-Desarrollador <a id='antigravity'></a>

### ¿Qué es Antigravity en el contexto de este proyecto?

Antigravity no es una herramienta pasiva que ejecuta instrucciones. Es el sistema de IA que ayudó a:

1. **Diseñar la arquitectura de skills** — proponer qué módulos crear y cómo organizarlos
2. **Generar el código base** — producir implementaciones funcionales que luego se revisan y refinan
3. **Mantener consistencia** — asegurar que los ejemplos en U5_01 sean coherentes con los de U5_07
4. **Detectar errores de diseño** — señalar cuando una decisión técnica crea deuda técnica
5. **Actualizar el contenido** — cuando salió Gemini 2.5, el contenido se actualizó en horas, no semanas

### El ciclo de colaboración

```
Humano: "Necesito una notebook sobre LangGraph con conexiones al dominio de la nanotecnología"
    ↓
IA: Propone estructura de secciones, identifica gaps de conocimiento del usuario
    ↓
Humano: Revisa la propuesta, ajusta objetivos, especifica nivel de dificultad
    ↓
IA: Genera el código, los ejemplos, la teoría
    ↓
Humano: Ejecuta el código, identifica qué no funciona o no queda claro
    ↓
IA: Itera y corrige
    ↓
[Ciclo continuo hasta calidad aprobada]
```

### ¿Quién tomó las decisiones importantes?

Esta es la pregunta más importante. La respuesta es: **el humano siempre**. El sistema de IA propuso opciones, evaluó trade-offs, y generó implementaciones. Pero las decisiones que definen el carácter del proyecto — qué se enseña, para quién, con qué nivel de profundidad — fueron siempre decisiones humanas.

Algunos ejemplos concretos:

| Decisión | Quién la tomó | Por qué importa |
|----------|---------------|-----------------|
| Usar Gemini Flash como modelo por defecto | Humano | Equilibrio costo-calidad para estudiantes |
| Incluir modelos locales (Ollama) como alternativa | Humano | Privacidad y acceso sin costo para estudiantes |
| Crear `external_skills/` como módulo reutilizable | Colaborativo | La IA propuso, el humano aprobó la dirección |
| Agregar DeepSeek y Qwen a la tabla de modelos | Humano | Visibilidad de alternativas de código abierto |
| Profundidad de las notebooks (6-8h vs 2-3h) | Humano | Decisión pedagógica del nivel del curso |

## 3. La Arquitectura de Skills: Decisiones y Consecuencias <a id='skills-architecture'></a>

### Alta decisión #1: Skills como módulos Python independientes

**Alternativa considerada:** Embeber las funciones de skills directamente en cada notebook.

**Decisión tomada:** Crear `external_skills/` como un paquete Python con módulos independientes.

**Consecuencias positivas:**
- Las skills se pueden importar desde cualquier notebook con `from external_skills.memory.episodic_retriever import retrieve`
- Los tests unitarios en `tests/unit_05/` prueban las skills de forma aislada
- Una skill mejorada en el módulo beneficia automáticamente todas las notebooks que la usan

**Consecuencias negativas:**
- Las notebooks no son 100% auto-contenidas — requieren que el paquete esté instalado
- Los paths relativos (`../../../../`) son frágiles si la estructura del repo cambia

**La lección:** En proyectos educativos, la auto-contención de notebooks facilita la distribución, pero la reutilización de código requiere módulos. El balance elegido aquí prioriza el patrón de software correcto sobre la conveniencia de distribución individual.

---

### Alta decisión #2: SHA-256 en la carga de skills remotas

**El problema encontrado:** La función `github_skill_loader.py` originalmente cargaba código Python desde GitHub usando `exec()` sin verificar la integridad del código descargado.

**El riesgo (OWASP A08):** Un atacante que comprometiera el repositorio de GitHub (o realizara un man-in-the-middle) podría ejecutar código arbitrario en el entorno del estudiante.

**La solución:** Agregar verificación SHA-256 antes de ejecutar cualquier skill descargada. El hash esperado se configura localmente, y si no coincide, la carga falla con `HashMismatchError`.

**La lección para estudiantes:** La IA puede generar código funcional rápidamente, pero los revisores humanos (o la propia IA en modo revisor) deben detectar vulnerabilidades de seguridad. En este proyecto, el código fue generado, revisado, y la vulnerabilidad fue identificada en el proceso de revisión — no en producción.

---

### Alta decisión #3: Versionado en el registro de skills

**Motivación:** A medida que el curso evolucionó, algunas skills cambiaron su interfaz. Una skill de memoria que en v1.0.0 recibía `content: str` empezó a recibir en v2.0.0 un dict con metadatos adicionales.

**La solución:** El `registry.py` implementa el patrón `skill@version` que permite coexistencia de múltiples versiones. Las notebooks más antiguas usan `skill@1.0.0` explícitamente; las nuevas usan la clave sin versión (→ latest).

**La lección:** Los proyectos educativos tienen ciclos de vida largos. Las decisiones de versionado tomadas al inicio determinan el costo de mantenimiento durante años.

## 4. Patrones de Colaboración Humano-IA <a id='human-ai-patterns'></a>

### Patrón 1: El Agente Arquitecto

Antes de generar cualquier código, se le pide al sistema de IA que **diseñe la arquitectura en alto nivel**. El output es un documento (como `PLAN_ACTUALIZACION_UNIDAD_5.md`) con:

- Objetivos de aprendizaje explícitos
- Lista de componentes a desarrollar
- Dependencias entre componentes
- Criterios de evaluación
- Errores comunes conocidos

Este documento actúa como **contrato** entre el humano y el sistema de IA. Antes de generar código, se acuerda qué se va a construir.

**Por qué funciona:** Evita construir el componente equivocado — el error más costoso en cualquier proyecto.

---

### Patrón 2: El Protocolo como Spec de Calidad

El `PROTOCOLO_UNIDAD_5_MULTIAGENTE.md` no es documentación — es una **especificación ejecutable de calidad**. Cada sección de una notebook debe tener:

- Warm-up con verificación de entorno
- Teoría antes del código
- Código con comentarios pedagógicos
- Output esperado documentado
- Conexión con la siguiente notebook

Cuando el sistema de IA genera una notebook, se le pasa el protocolo como contexto y se le instruye a verificar cada componente antes de entregarlo.

**Por qué funciona:** Las specs explícitas reducen la variabilidad entre notebooks generadas en sesiones distintas.

---

### Patrón 3: El Ciclo de Scoring

Antes de generar el contenido final, se evalúa cada componente en 5 dimensiones:
- Actualidad de tecnologías (0-100)
- Optimalidad del código (0-100)
- Operatividad (¿el código realmente funciona?) (0-100)
- Integración con vibe coding (0-100)
- Calidad didáctica (0-100)

Si alguna dimensión está por debajo de 90, se generan las correcciones específicas antes de proceder.

**Por qué funciona:** El proceso de scoring fuerza al sistema de IA a articular deficiencias que de otra forma quedarían implícitas.

---

### Patrón 4: Handoff con Contexto Completo

Cada sesión de trabajo con el sistema de IA termina con un resumen de sesión que incluye:
- Qué se completó
- Qué está en progreso y en qué punto exacto
- Qué está pendiente y en qué orden

Cuando se retoma el trabajo en una nueva sesión, el primer paso es cargar ese contexto.

**Por qué funciona:** Los LLMs no tienen memoria persistente entre sesiones. El contexto explícito es lo más parecido a "donde quedamos".

**"Hazlo más simple"** — a veces simplifió correctamente, a veces eliminó conceptos pedagógicamente importantes. La instrucción correcta es: "simplifica sin perder los conceptos X, Y, Z".

## 6. Lessons Learned: Lo que Funcionó y lo que No <a id='lessons'></a>

### Lo que funcionó excepcionalmente bien

**1. Desarrollo incremental con documentos de plan:** Tener `PLAN_ACTUALIZACION_UNIDAD_5.md` como fuente de verdad significó que cualquier colaborador humano o IA pudiera retomar el trabajo desde cualquier punto sin perder contexto.

**2. El registro de skills como interfaz estable:** Al definir primero la interfaz de las skills (qué parámetros reciben, qué retornan) y luego implementarlas, fue posible escribir tests antes que el código — práctica de TDD aplicada a desarrolló con IA.

**3. Modelos locales como fallback universal:** Cada componente del sistema tiene un fallback a Ollama. Esto hizo posible que estudiantes sin API keys pudieran ejecutar el 80% del material.

**4. Documentación de errores comunes en el protocolo:** Al documentar explícitamente los errores que los estudiantes cometen (en la sección de criterios por notebook), el sistema de IA los evitó en el código generado.

---

### Lo que no funcionó (y cómo se corrigió)

**1. Generar código sin ejecutarlo primero**
- **Problema:** Algunas celdas tenían errores de importación que solo aparecen al ejecutar.
- **Corrección:** El protocolo ahora especifica que toda celda de código debe incluir un bloque `expected_output` documentando qué debería imprimir cuando funciona correctamente.

**2. Inconsistencia de modelos entre notebooks**
- **Problema:** U5_01 usaba `gpt-4` y U5_04 usaba `gemini-pro` — distintas denominaciones.
- **Corrección:** Tabla de modelos canónica en el PLAN con nombres exactos y versiones pinned.

**3. Skills sin tests**
- **Problema:** Una skill de routing tenía un bug silencioso que solo aparecía con inputs específicos.
- **Corrección:** `tests/unit_05/test_skills.py` con 29 tests que cubren edge cases conocidos. El protocolo ahora exige tests antes de considerar una skill completa.

**4. Dependencias implícitas entre notebooks**
- **Problema:** U5_05 asumía que ChromaDB ya estaba instalado (instalado en U5_02) sin verificarlo.
- **Corrección:** Cada notebook verifica independientemente sus dependencias en el warm-up.

Esta decisión prioriza la **distribuibilidad** (un estudiante puede copiar solo la carpeta `Proyecto final/` y todo funciona) sobre la **reutilización** (los tools no se pueden importar desde otros proyectos sin copiar código).

### Lo que este proyecto demuestra

Un humano con conocimiento de dominio (nanotecnología, pedagogía, diseño de cursos) y un sistema de IA con capacidades de generación de código pueden construir, juntos, contenido educativo que ninguno podría producir solo en el mismo tiempo.

- Velocidad de síntesis (producir primeras versiones en minutos, no días)

- Detección de errores (revisar código por vulnerabilidades que el humano no vería)

- **Entender el contexto institucional** (qué puede pedirse en un examen, qué equipos tienen los estudiantes).
