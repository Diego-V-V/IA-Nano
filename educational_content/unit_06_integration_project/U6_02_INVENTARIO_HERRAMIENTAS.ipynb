```

        │                               │
        ▼                               ▼
[U3] Grad-CAM                  [U5] Agente Medidor (image processing)
        │                               │
        └──────────────┬────────────────┘
                       ▼
              [U5] Agente Científico (LLM via OpenRouter)
                       │
              [U5] RAG - ChromaDB
                       │
                       ▼
              [U5] Agente Reportero
                       │
                       ▼
              [U6] FastAPI REST
```

## Plan Técnico del Proyecto SEM

El pipeline del proyecto fluye en **7 etapas secuenciales**. Cada etapa tiene una herramienta del curso que la implementa:

| Etapa | Descripción | Unidad |
|---|---|---|
| 1. Adquisición | Descarga del NFFA-EUROPE SEM Dataset | U4 |
| 2. Preprocesamiento | Resize 224×224, normalización ImageNet, Data Augmentation | U3 |
| 3. Entrenamiento | Transfer Learning ResNet-18 (10 épocas FC + 10 épocas FT) | U3 |
| 4. Interpretabilidad | Grad-CAM sobre `layer4[-1]`, análisis de activaciones | U3 |
| 5. Multi-Agente | Orquestación de 5 agentes con LangGraph | U5 |
| 6. RAG | ChromaDB con literatura SEM para fundamentar hallazgos | U5 |
| 7. Despliegue | API REST FastAPI con endpoints `/analyze` y `/health` | U6 |

La siguiente celda define este plan técnico y lo exporta como JSON.

````python
# ============================================================
#   DEFINE TU PIPELINE TECNICO
#   Describe cada etapa de tu proyecto y la herramienta que usas
# ============================================================

mi_pipeline = [
    # Formato: (etapa, descripcion_lo_que_hace, herramienta_del_catalogo)
    ("Etapa 1: Adquisicion", "Descarga y organización del NFFA-EUROPE SEM Dataset", "U4_analisis_datos_exp"),
    ("Etapa 2: Preprocesamiento", "Resize, normalización ImageNet y data augmentation de imágenes", "U3_redes_neuronales"),
    ("Etapa 3: Entrenamiento", "Transfer Learning sobre ResNet-18 (FC 10 epochs + FT 10 epochs)", "U3_redes_neuronales"),
    ("Etapa 4: Interpretabilidad", "Cálculo de métricas y evaluación visual con Grad-CAM", "U3_redes_neuronales"),
    ("Etapa 5: Sistema Multi-Agente", "Orquestación de 5 agentes para análisis (Clasificador, Medidor, Visualizador, Científico, Reportero)", "U5_agentes_langchain"),
    ("Etapa 6: RAG y Literatura", "Búsqueda semántica de literatura científica en ChromaDB para fundamentación", "U5_rag_memoria"),
    ("Etapa 7: Despliegue API", "Exposición del pipeline multi-agente en una API REST", "U6_api_fastapi")
]

# Validar que no quedaron placeholders
pipeline_valido = all(
    etapa[1] not in ("...", "") and etapa[2] not in ("...", "")
    for etapa in mi_pipeline
)

print("PIPELINE TECNICO DE TU PROYECTO")
print("-" * 55)
for i, (nombre, descripcion, herramienta) in enumerate(mi_pipeline, 1):
    print(f"  Paso {i}: {nombre}")
    print(f"    Hace     : {descripcion}")
    print(f"    Herramienta: {herramienta}")

print()
if pipeline_valido:
    print("Pipeline listo. Continua con U6_03_IMPLEMENTACION_PROYECTO.ipynb")
else:
    print("Completa todos los campos '...' del pipeline antes de continuar.")
````
