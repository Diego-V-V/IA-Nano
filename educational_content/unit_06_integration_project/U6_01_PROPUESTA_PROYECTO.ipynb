## ¿De qué trata este proyecto?

Este proyecto integra las seis unidades del curso para construir un **sistema inteligente de análisis de imágenes SEM** (Microscopía Electrónica de Barrido) de nanomateriales.

El pipeline completo:
1. **Datos reales** — NFFA-EUROPE SEM Dataset con 7,068 micrografías etiquetadas
2. **Modelo CNN** — ResNet-18 con Transfer Learning desde ImageNet → **97.74% accuracy**
3. **Interpretabilidad** — Grad-CAM sobre `layer4[-1]` para visualizar qué aprende el modelo
4. **Multi-Agente** — 5 agentes coordinados: Clasificador, Medidor, Visualizador, Científico, Reportero
5. **RAG** — ChromaDB con literatura de nanotecnología para fundamentar los resultados
6. **API REST** — FastAPI que expone el pipeline completo al mundo

## ¿Por qué es importante?

La clasificación manual de micrografías SEM en laboratorios es **lenta, tediosa y propensa a errores humanos**. Automatizar este proceso con IA permite acelerar el control de calidad en la síntesis de nanomateriales, lo que tiene impacto directo en industrias como energía, medicina y materiales avanzados.
