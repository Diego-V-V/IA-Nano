# Prueba del Sistema Multi-Agente — Clasificador SEM
## Verificación de que los 4 agentes funcionan correctamente

Este notebook ejecuta cada agente del sistema multi-agente de la NB04 de forma
independiente para verificar que:
1. El **Analista** interpreta correctamente las métricas del modelo
2. El **Científico** valida la interpretabilidad de Grad-CAM
3. El **Escritor** genera un reporte científico estructurado
4. El **Revisor** evalúa la calidad del reporte con un score numérico

**LLM:** OpenRouter (GPT-4o-mini / fallback automático)

---

```python
# ============================================================
# CONFIGURACION: API key y metricas del modelo
# ============================================================
import os, json, time, re
from pathlib import Path

os.environ["OPENROUTER_API_KEY"] = "TU_OPENROUTER_API_KEY"

PROJECT_ROOT = Path('..').resolve()
with open(PROJECT_ROOT / 'reports' / 'model_metrics.json', encoding='utf-8') as f:
    MODEL_METRICS = json.load(f)

print("Metricas del modelo cargadas:")
print(f"  Modelo:    {MODEL_METRICS['model']}")
print(f"  Tarea:     {MODEL_METRICS['task']}")
print(f"  Accuracy:  {MODEL_METRICS['accuracy']:.4f} ({MODEL_METRICS['accuracy']:.1%})")
print(f"  F1-score:  {MODEL_METRICS['f1_score']:.4f}")
print(f"  Precision: {MODEL_METRICS['precision']:.4f}")
print(f"  Recall:    {MODEL_METRICS['recall']:.4f}")
print(f"  AUC-ROC:   {MODEL_METRICS['auc_roc']:.4f}")
print(f"  Errores:   {MODEL_METRICS['n_errors']} de {MODEL_METRICS['test_size']}")
print(f"  Grad-CAM:  {MODEL_METRICS['gradcam_interpretation']}")
```

```python
# ============================================================
# CONECTAR LLM via OpenRouter
# ============================================================
from langchain_openai import ChatOpenAI

OPENROUTER_API_KEY = os.environ["OPENROUTER_API_KEY"]

MODELS_FALLBACK = [
    "openai/gpt-4o-mini",
    "anthropic/claude-3-haiku",
    "google/gemini-2.5-flash",
]

llm = None
OPENROUTER_MODEL = ""
for model_id in MODELS_FALLBACK:
    try:
        _c = ChatOpenAI(
            base_url="https://openrouter.ai/api/v1",
            api_key=OPENROUTER_API_KEY,
            model=model_id,
            temperature=0,
            default_headers={
                "HTTP-Referer": "https://github.com/antigravity-nano",
                "X-Title": "SEM Classifier Test",
            },
        )
        _c.invoke("OK")
        llm = _c
        OPENROUTER_MODEL = model_id
        print(f"LLM conectado exitosamente: {model_id}")
        break
    except Exception as e:
        print(f"  [!] {model_id}: {e}")

if llm is None:
    raise RuntimeError("No se pudo conectar a ningun modelo OpenRouter.")
```

## Tools: Acceso a métricas y búsqueda de papers

```python
# ============================================================
# TOOLS para los agentes
# ============================================================
from langchain_core.tools import tool
import httpx
import xml.etree.ElementTree as ET

@tool
def load_model_metrics() -> str:
    """Carga las metricas del clasificador SEM entrenado."""
    m = MODEL_METRICS
    return (
        f"Modelo: {m['model']}\n"
        f"Tarea: {m['task']}\n"
        f"Dataset: {m['dataset']} ({m['test_size']} imagenes test)\n"
        f"Accuracy: {m['accuracy']:.4f} ({m['accuracy']:.1%})\n"
        f"F1-score: {m['f1_score']:.4f}\n"
        f"Precision: {m['precision']:.4f}\n"
        f"Recall: {m['recall']:.4f}\n"
        f"AUC-ROC: {m['auc_roc']:.4f}\n"
        f"Errores: {m['n_errors']} ({m['error_rate']:.1%})\n"
        f"Clases: {m['classes']}\n"
        f"Matriz de confusion: {m['confusion_matrix']}\n"
        f"Grad-CAM: {m['gradcam_interpretation']}"
    )

@tool
def search_sem_papers(query: str, max_results: int = 3) -> str:
    """Busca papers cientificos en arXiv sobre clasificacion SEM."""
    url = "http://export.arxiv.org/api/query"
    params = {"search_query": f"all:{query}", "max_results": max_results, "sortBy": "relevance"}
    try:
        response = httpx.get(url, params=params, timeout=15)
        response.raise_for_status()
        root = ET.fromstring(response.text)
        ns = {"atom": "http://www.w3.org/2005/Atom"}
        results = []
        for entry in root.findall("atom:entry", ns):
            title = entry.find("atom:title", ns).text.strip()
            abstract = entry.find("atom:summary", ns).text.strip()[:200]
            results.append(f"* {title}\n  {abstract}")
        return "\n\n".join(results) if results else "Sin resultados."
    except Exception as e:
        return f"Error buscando papers: {e}"

print("Tools definidos: load_model_metrics, search_sem_papers")
```

---
## Agente 1: Analista de Métricas

El analista carga las métricas del modelo, busca papers relevantes en arXiv,
y produce un resumen comparativo con la literatura existente.

```python
# ============================================================
# AGENTE 1: ANALISTA DE METRICAS
# ============================================================
print("Cargando metricas del modelo...")
metricas = load_model_metrics.invoke({})
print(metricas)

print("\n" + "="*50)
print("Buscando papers relevantes en arXiv...")
papers = search_sem_papers.invoke({"query": "SEM image classification nanoparticles nanowires CNN", "max_results": 3})
print(papers)
```

```python
# ============================================================
# ANALISIS DEL ANALISTA (LLM)
# ============================================================
t0 = time.time()

analyst_prompt = f"""Eres un analista de ML especializado en Computer Vision para microscopia SEM.

Metricas del modelo:
{metricas}

Papers relevantes:
{papers}

Produce un resumen analitico de 200-300 palabras:
1. Evalua si el accuracy (99.07%), F1 (99.03%) y AUC-ROC (99.43%) son competitivos
2. Compara con los resultados reportados en los papers
3. Comenta que el Grad-CAM muestra features fisicamente relevantes
4. Identifica fortalezas y debilidades del modelo"""

print("Enviando al LLM...")
analyst_response = llm.invoke(analyst_prompt)
metrics_summary = analyst_response.content
elapsed = time.time() - t0
print(f"Respuesta recibida en {elapsed:.1f}s")
```

```python
# Mostrar respuesta del Analista con formato Markdown
from IPython.display import Markdown, display

display(Markdown("### Respuesta del Agente Analista"))
display(Markdown(metrics_summary))
```

---
## Agente 2: Científico SEM

El científico evalúa si los resultados tienen sentido físico:
- ¿Las métricas son aceptables para clasificación de morfologías SEM?
- ¿El Grad-CAM detecta features físicamente relevantes?
- ¿Qué limitaciones y mejoras futuras existen?

```python
# ============================================================
# AGENTE 2: CIENTIFICO SEM
# ============================================================
t0 = time.time()

scientist_prompt = f"""Eres un cientifico experto en microscopia SEM y nanomateriales.

Analisis previo del modelo:
{metrics_summary[:800]}

Evalua:
1. Son las metricas del modelo aceptables para clasificacion de morfologias SEM?
2. El Grad-CAM muestra que el modelo aprende features fisicamente relevantes?
   (bordes redondeados para nanoparticulas, estructuras elongadas para nanohilos)
3. Que limitaciones tiene este enfoque?
4. Que mejoras sugieres para investigacion futura?

Genera un analisis cientifico estructurado de 200-300 palabras."""

print("Enviando al LLM...")
scientist_response = llm.invoke(scientist_prompt)
scientific_analysis = scientist_response.content
elapsed = time.time() - t0
print(f"Respuesta recibida en {elapsed:.1f}s")
```

```python
# Mostrar respuesta del Cientifico
display(Markdown("### Respuesta del Agente Cientifico"))
display(Markdown(scientific_analysis))
```

---
## Agente 3: Escritor de Reporte Científico

El escritor toma los análisis de los agentes anteriores y genera un reporte
científico completo en Markdown con 7 secciones.

```python
# ============================================================
# AGENTE 3: ESCRITOR DE REPORTE
# ============================================================
t0 = time.time()

writer_prompt = f"""Escribe un reporte cientifico completo en Markdown sobre el clasificador SEM.

DATOS DEL ANALISIS:
{metrics_summary[:600]}

VALIDACION CIENTIFICA:
{scientific_analysis[:600]}

ESTRUCTURA REQUERIDA:
## Resumen Ejecutivo
## Metodologia
## Resultados y Metricas
## Interpretabilidad (Grad-CAM)
## Comparacion con Literatura
## Limitaciones y Trabajo Futuro
## Conclusiones

Extension: 400-600 palabras. Formato Markdown. Lenguaje academico pero accesible."""

print("Enviando al LLM (esta respuesta es mas larga, puede tardar ~15s)...")
writer_response = llm.invoke(writer_prompt)
draft_report = writer_response.content
elapsed = time.time() - t0
print(f"Reporte generado en {elapsed:.1f}s")
```

```python
# Mostrar el reporte completo del Escritor
display(Markdown("### Reporte generado por el Agente Escritor"))
display(Markdown(draft_report))
```

---
## Agente 4: Revisor

El revisor evalúa la calidad del reporte con un score de 0-100 y proporciona
feedback específico. Si el score es ≥ 75, el reporte se aprueba.

```python
# ============================================================
# AGENTE 4: REVISOR
# ============================================================
t0 = time.time()

reviewer_prompt = f"""Evalua este reporte cientifico sobre clasificacion SEM.

REPORTE:
{draft_report[:2000]}

Responde EXACTAMENTE con este formato:
SCORE: [numero 0-100]
FEEDBACK: [comentario detallado sobre que esta bien, que falta, y sugerencias de mejora]"""

print("Enviando al LLM...")
reviewer_response = llm.invoke(reviewer_prompt)
review_text = reviewer_response.content.strip()
elapsed = time.time() - t0

# Extraer score
score_match = re.search(r'SCORE:\s*(\d+)', review_text)
score = int(score_match.group(1)) if score_match else 0

print(f"Revision completada en {elapsed:.1f}s")
```

```python
# Mostrar la revision
display(Markdown("### Evaluacion del Agente Revisor"))
display(Markdown(review_text))

# Veredicto visual
if score >= 75:
    display(Markdown(f"## \u2705 REPORTE APROBADO (Score: {score}/100)"))
else:
    display(Markdown(f"## \u274c REPORTE NECESITA REVISION (Score: {score}/100)"))
```

---
## Resumen Final

```python
# ============================================================
# RESUMEN FINAL
# ============================================================
print("=" * 60)
print("RESUMEN DE LA PRUEBA MULTI-AGENTE")
print("=" * 60)
print(f"  LLM usado:     {OPENROUTER_MODEL}")
print(f"  Analista:      Genero resumen comparativo con literatura")
print(f"  Cientifico:    Valido interpretabilidad de Grad-CAM")
print(f"  Escritor:      Genero reporte Markdown de 7 secciones")
print(f"  Revisor:       Score = {score}/100")
print(f"  Veredicto:     {'APROBADO' if score >= 75 else 'NECESITA REVISION'}")
print()

# Guardar reporte
report_path = PROJECT_ROOT / 'reports' / 'test_agent_report.md'
with open(report_path, 'w', encoding='utf-8') as f:
    f.write(f"# Reporte generado por sistema multi-agente\n")
    f.write(f"**Modelo LLM:** {OPENROUTER_MODEL}  \n")
    f.write(f"**Score del revisor:** {score}/100\n\n---\n\n")
    f.write(draft_report)
print(f"  Reporte guardado en: {report_path}")
print(f"\nPrueba completada exitosamente.")
```
